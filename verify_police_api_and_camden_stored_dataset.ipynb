<h1>A Sample Month's Dataset Comparison between the Camden Dataset and Police API Retreived one</h1>
This notebook is just intended to check a sample retrieved from Police API against the same offline (downloaded) Camden dataset. An ideal situation would be that they match perfectly. We used June, 2019 month's data as a sample here. Some conclusions can be found at the end of this notebook.

```python
# python libraries: pandas, numpy and datetime
import numpy as np
import pandas as pd

from datetime import datetime
```

```python
# this is the dataset retrieved through Police API
df1 = pd.read_csv('data/Police_API_Sample_July_2019.csv')
df1['Epoch'] = df1['Epoch'].apply(lambda x: datetime.strptime(x, '%m/%d/%Y %I:%M:%S %p'))
df1= df1.set_index(['Epoch'])
df1.head()
```

```python
# this is the dataset from the Camden CSV file
df = pd.read_csv('data/On_Street_Crime_in_Camden.csv')
df['Epoch'] = df['Epoch'].apply(lambda x: datetime.strptime(x, '%m/%d/%Y %I:%M:%S %p'))
df= df.set_index(['Epoch'])
df.head()
```

```python
print('June 2019 Dataset Size (Camden CSV file):{}\nJune 2019 Dataset Size (Police API):{}'.format(df.resample('M').count().loc['2019-06-30']['Category'], df1.resample('M').count()['Category'].values[0]))

d1 = dict(zip(list(df1.groupby(['Category']).indices.keys()), list(df1.groupby(['Category']).size())))
d2 = dict(zip(list(df1.groupby(['Category']).indices.keys()), list(df.loc['2019-06'].groupby(['Category']).size())))
df2 = pd.DataFrame([d1, d2], index=['Police API', 'Camden CSV'])
df2.head()
```

```python
# already some inconsistencies as can be seen from the above analysis
# we are outlining one particular Category 'Possession of weapons' records to compare both the dataset
# this one from the CSV dataset
df[df['Category']=='Possession of weapons'].loc['2019-06'].sort_values(by=['ID'])
```

```python
# this is from the Police API
df1[df1['Category']=='Possession of weapons'].sort_values(by=['ID'])
```

<h1>Conclusion</h1>


Even though majority of the crime data information inside both dataset match, there are some inconsistencies as well. This is revealed here using a sample [June 2019 data from both dataset]. Camden dataset is not maintained by the borough while the date from Police API is updated regularly. There are certainly some records [especially for the crime outcome] which were not updated inside the Camden dataset.

More investigation might be required to explain the minor inconsistencies. If given an option, the Police API dataset might be more suitable since they are maintained and updated regularly.
